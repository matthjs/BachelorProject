```python
import numpy as np

def generate_nd_grid(ranges, num_points):
    grids = [np.linspace(r[0], r[1], num_points) for r in ranges]
    mesh = np.meshgrid(*grids)
    grid_points = np.vstack([m.ravel() for m in mesh]).T
    return grid_points

# Example usage
n_dims = 3
ranges = [(0, 1)] * n_dims  # Assuming unit cube for simplicity
num_points = 100
grid_points = generate_nd_grid(ranges, num_points)

print(grid_points.shape)
```

```python
from matplotlib import pyplot as plt

# Plotting 3D points
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = grid_points[:, 0]
y = grid_points[:, 1]
z = grid_points[:, 2]

ax.scatter(x, y, z, c='b', marker='o')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
# Choose dimensions to plot (e.g., dimensions 2 and 4)
dim_i = 1  # Second dimension (0-based index)
dim_j = 2  # Fourth dimension (0-based index)

# Extract points from the chosen dimensions
points_2d = grid_points[:, [dim_i, dim_j]]
```

```python
# Plotting 2D points
plt.figure()
plt.scatter(points_2d[:, 0], points_2d[:, 1], c='b', marker='o')
plt.xlabel(f'Dimension {dim_i + 1}')
plt.ylabel(f'Dimension {dim_j + 1}')
plt.title(f'Points from Dimension {dim_i + 1} and {dim_j + 1}')
plt.show()
```
